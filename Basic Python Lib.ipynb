```python
gb= df.groupby("sex").agg({"G1":"mean", "G2":"mean", "G3":"mean"})
print(gb)
```

```python
at_home=df.loc[(df['Mjob']=="at_home")].count()
health=df.loc[(df['Mjob']=="health")].count()
other=df.loc[(df['Mjob']=="other")].count()
print(at_home, health, other)
```

```python
l=["health", "at_home", "other"]
mlist=[at_home["Mjob"],health["Mjob"], other["Mjob"]]
plt.pie(mlist, labels=l,autopct="%1.1f%%")
plt.show()
```
